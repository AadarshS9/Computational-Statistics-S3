```python
print('Program to use DataFrame.transform() function to add 10 to each element in the dataframe\n')
import pandas as pd
df = pd.DataFrame({"A":[12, 4, 5, None, 1],"B":[7, 2, 54, 3, None],"C":[20, 16, 11, 3, 8],"D":[14, 3, None, 2, 6]})
index_ = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5']
df.index = index_
print(df)
print("\n")
#add 10 to each element of the dataframe
result = df.transform(func = lambda x : x + 10)
print(result)
```

```python
print('Program to use Pandas.pivot() function to create a pivot table\n')
import pandas as pd
df = pd.DataFrame({'A': ['John', 'Boby', 'Mina'],'B': ['Masters', 'Graduate', 'Graduate'],'C': [27, 23, 21]})
print(df)
print("\n")
# values can be an object or a list
df.pivot('A', 'B', 'C')
```

```python
print('Program to use Pandas.pivot() function to create a pivot table\n')
import pandas as pd
df = pd.DataFrame({'A': ['John', 'Boby', 'Mina'],'B': ['Masters', 'Graduate', 'Graduate'],'C': [27, 23, 21]})
# value is a list
df.pivot(index ='A', columns ='B', values =['C', 'A'])
```

```python
print('Program to iterate over characters of a string in Python\n')
string_name = 'Malayalam'
print('...1)Iterate over the string\n')
for element in string_name:
  print(element, end=' ')
  #print("\n")

print("\n")
print('...2)Iterate over the string\n')
for element in string_name:
  print(element, end=' ')
  print("\n")

print('...3)Iterate over the index\n')
for element in range(0, len(string_name)):
  print(string_name[element])
```

```python
print('Program to iterate over characters of a string in reverse order in Python\n')
string_name = 'Malayalam'
print('...1)Slicing the string in reverse fashion\n')
for element in string_name[ : :-1]:
  print(element, end =' ')
  print('\n')

print('...2)Slicing the string in reverse fashion\n')
for element in string_name[ : :-1]:
  print(element, end =' ')
print('\n')


print('...3)Using reversed() function\n')
ran = len(string_name)
for element in reversed(range(0, ran)):
  print(string_name[element])
```

```python
print('Program to iterate over a particular set of element in Python\n')
string_name = 'malayalam'
# string_name[start:end:step]
for element in string_name[2:8:1]:
  print(element, end =' ')
```
